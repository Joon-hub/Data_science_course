### Seaborn: distplot

```python
from matplotlib import pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins')

penguins.shape
```

```python
penguins.head()
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

#### Basics

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=5);
```

```python
sns.set_style('darkgrid')
sns.histplot(y = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=5);
```

```python
# making plot more granualr using binwidth and binrange

sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=20,
             binrange=(180,220));
```

'''stat'''

```python
# change yaxis to count
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=10,
             stat = 'count');

```

```python
# change yaxis to density
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=10,
             stat = 'density');
```

```python
# probability height of histigram bars equals to 1
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = True,
             binwidth=10,
             stat = 'probability');
```

```python
# probability height of histigram bars equals to 1
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=10,
             stat = 'probability',
             element='step',
             fill = True);
```

**Cummulativ**

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             stat='probability', 
             fill=False, 
             element='step',
             cumulative=True
            );
```

**Hue**

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=5,
             hue = 'species');
```

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=5,
             hue = 'species',
             element='poly');
```

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=5,
             hue = 'species',
             element='step');
```

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=5,
             hue = 'species',
             multiple='stack');
```

```python
sns.set_style('darkgrid')
sns.histplot(x = 'flipper_length_mm', data = penguins,
             kde = False,
             binwidth=5,
             hue = 'species',
             multiple='fill');
```

***BONUS Example:*** multiple="dodge"

```python
sns.histplot(x='species', data=penguins, 
             hue='island', 
             multiple='dodge',
             shrink=0.8
            ); 
```

```python
#platette
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins,
             shrink=0.8, palette='bone'
            );
```

```python
# fill - flase
sns.histplot(x='flipper_length_mm', data=penguins,
            fill=False
            );
```
