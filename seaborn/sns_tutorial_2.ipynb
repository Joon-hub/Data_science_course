### Seaborn: KDEplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

**Intro Visuals**

### 1. Univariate 

```python
sns.kdeplot(cars.horsepower,fill = True);
```

```python
sns.kdeplot(cars.horsepower,fill = True, color = 'pink' ,bw_method = 0.05);
```

```python
sns.kdeplot(cars.horsepower,cumulative = True, fill = False, color = 'green' ,bw_method = 0.05);
```

### 2. Bivariate 

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, cmap="viridis");
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, 
            cmap="viridis", 
            n_levels = 20,
            cbar = True);
# n_levels add more graduality
```

#### 2.1 Bivaraite plot by category

```python
cyl_4 = cars[cars.cylinders== 4]
cyl_8 = cars[cars.cylinders== 8]

plt.figure(figsize=(8,4))
sns.kdeplot(x = cyl_4.horsepower,y = cyl_4.mpg, cmap = "Blues", thresh=0.05,fill= True)
sns.kdeplot(x = cyl_8.horsepower,y = cyl_8.mpg, cmap ="Reds", thresh= 0.05, fill = True)

plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Miles per Gallon (MPG)', fontsize=14)

plt.annotate("4 Cylinders", (105, 32), color='blue', fontsize=16, fontweight='bold')
plt.annotate("8 Cylinders", (190, 18), color='red', fontsize=16, fontweight='bold');
plt.show()
```
